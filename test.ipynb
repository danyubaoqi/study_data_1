将reg_preference_for_trad转为数字

```python
dic={"一线城市":1,"二线城市":2,"三线城市":3,"境外":4,"其他城市":5}
rd['reg_preference_for_trad'].value_counts()
for i in range(0, rd.shape[0]):
    rd.reg_preference_for_trad[i] = dic[rd.reg_preference_for_trad[i]]
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg' # 矢量图
plt.figure(figsize=(35, 20))
count = 1
for i in rd.columns.values:
    try:
        plt.subplot(8,11,count)
        rd[i].plot(kind = 'box')
        count += 1
    except:
        pass
```

处理缺失值

```python
rd = rd.dropna(axis=0, thresh = rd.shape[1]*0.8) # 行
rd = rd.dropna(axis=1, thresh = rd.shape[0]*0.8) # 列
# 使用均值替换缺失值
data = rd.fillna(rd.mean())
data
```
